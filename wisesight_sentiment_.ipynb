##Step4: Data cleaning

```python
import re
import pandas as pd

def clean_text(example):
    text = example['texts']  # à¸ªà¸¡à¸¡à¸•à¸´à¸§à¹ˆà¸²à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¸„à¸·à¸­ 'texts'

    # à¸¥à¸š URLs
    text = re.sub(r'http\S+', '', text)

    # à¸¥à¸šà¸­à¸±à¸à¸‚à¸£à¸°à¸—à¸µà¹ˆà¹„à¸¡à¹ˆà¹ƒà¸Šà¹ˆà¸ à¸²à¸©à¸²à¹„à¸—à¸¢à¹à¸¥à¸°à¸Šà¹ˆà¸­à¸‡à¸§à¹ˆà¸²à¸‡
    text = re.sub(r'[^à¸-à¹™\s]', '', text)

    # à¸¥à¸šà¸Šà¹ˆà¸­à¸‡à¸§à¹ˆà¸²à¸‡à¸‹à¹‰à¸³à¸à¸±à¸™à¹à¸¥à¸°à¸¥à¸šà¸Šà¹ˆà¸­à¸‡à¸§à¹ˆà¸²à¸‡à¸•à¹‰à¸™/à¸—à¹‰à¸²à¸¢
    text = re.sub(r'\s+', ' ', text).strip()

    # à¸ªà¹ˆà¸‡à¸à¸¥à¸±à¸šà¸„à¹ˆà¸²à¸—à¸µà¹ˆà¹à¸à¹‰à¹„à¸‚
    example['texts'] = text
    return example

# à¹ƒà¸Šà¹‰ map à¸à¸±à¸š dataset à¸‚à¸­à¸‡à¸„à¸¸à¸“à¹€à¸žà¸·à¹ˆà¸­à¸—à¸³à¸„à¸§à¸²à¸¡à¸ªà¸°à¸­à¸²à¸”à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¹ƒà¸™à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'texts'
dataset['train'] = dataset['train'].map(clean_text)

#remove emoji
def remove_emojis(example):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emoticons
                               u"\U0001F300-\U0001F5FF"  # Symbols & Pictographs
                               u"\U0001F680-\U0001F6FF"  # Transport & Map Symbols
                               u"\U0001F1E0-\U0001F1FF"  # Flags
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251"  # Enclosed characters
                               "]+", flags=re.UNICODE)
    # à¹à¸—à¸™à¸—à¸µà¹ˆà¸­à¸´à¹‚à¸¡à¸ˆà¸´à¸”à¹‰à¸§à¸¢à¸„à¹ˆà¸²à¸§à¹ˆà¸²à¸‡
    example['train'] = emoji_pattern.sub(r'', example['train'])
    return example

# à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸¥à¸šà¸Šà¹ˆà¸­à¸‡à¸§à¹ˆà¸²à¸‡à¸‹à¹‰à¸³à¹à¸¥à¸°à¸ªà¹ˆà¸‡à¸„à¸·à¸™à¹€à¸›à¹‡à¸™ dict
def remove_duplicate_spaces(example):
    text = example['texts']
    text = re.sub(r'\s+', ' ', text)  # à¹à¸—à¸™à¸—à¸µà¹ˆà¸Šà¹ˆà¸­à¸‡à¸§à¹ˆà¸²à¸‡à¸«à¸¥à¸²à¸¢à¸•à¸±à¸§à¸”à¹‰à¸§à¸¢ 1 à¸Šà¹ˆà¸­à¸‡à¸§à¹ˆà¸²à¸‡
    return {'texts': text.strip()}  # à¸•à¸±à¸”à¸Šà¹ˆà¸­à¸‡à¸§à¹ˆà¸²à¸‡à¸—à¸µà¹ˆà¸­à¸¢à¸¹à¹ˆà¸•à¹‰à¸™à¹à¸¥à¸°à¸—à¹‰à¸²à¸¢à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡ à¹à¸¥à¹‰à¸§à¸ªà¹ˆà¸‡à¸„à¸·à¸™à¹€à¸›à¹‡à¸™ dict

# à¹ƒà¸Šà¹‰ map à¹€à¸žà¸·à¹ˆà¸­à¹à¸›à¸¥à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥
dataset['train'] = dataset['train'].map(remove_duplicate_spaces)
```

##Tokenization

```python
!pip install pythainlp
```

```python
from pythainlp.tokenize import word_tokenize

def tokenize_text(example):
    text = example['texts']  # à¸ªà¸¡à¸¡à¸•à¸´à¸§à¹ˆà¸²à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¸„à¸·à¸­ 'texts'
    tokens = word_tokenize(text, engine='newmm')  # à¹ƒà¸Šà¹‰ newmm tokenizer à¸‚à¸­à¸‡ pythainlp
    example['texts'] = ' '.join(tokens)  # à¸£à¸§à¸¡à¸„à¸³à¸—à¸µà¹ˆà¸•à¸±à¸”à¹à¸¥à¹‰à¸§à¹ƒà¸«à¹‰à¹€à¸›à¹‡à¸™à¸›à¸£à¸°à¹‚à¸¢à¸„à¹ƒà¸«à¸¡à¹ˆ
    return example

dataset['train'] = dataset['train'].map(tokenize_text)
```

##Normalization

```python
# à¸•à¸±à¸§à¸­à¸¢à¹ˆà¸²à¸‡à¸à¸²à¸£à¸¥à¸šà¸§à¸£à¸£à¸“à¸¢à¸¸à¸à¸•à¹Œ
from pythainlp.util import normalize

def normalize_text(example):
    text = example['texts']  # à¸ªà¸¡à¸¡à¸•à¸´à¸§à¹ˆà¸²à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¸„à¸·à¸­ 'texts'
    normalized_text = normalize(text)  # Normalize à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¹ƒà¸«à¹‰à¹€à¸›à¹‡à¸™à¸¡à¸²à¸•à¸£à¸à¸²à¸™
    return {'texts': normalized_text}  # à¸ªà¹ˆà¸‡à¸„à¸·à¸™à¹€à¸›à¹‡à¸™ dict

dataset['train'] = dataset['train'].map(normalize_text)
```

##Text Vectorization

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# à¸ªà¸£à¹‰à¸²à¸‡ tokenizer à¹à¸¥à¸°à¹à¸›à¸¥à¸‡à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¹€à¸›à¹‡à¸™à¸¥à¸³à¸”à¸±à¸šà¸‚à¸­à¸‡à¸•à¸±à¸§à¹€à¸¥à¸‚

texts_train = dataset['train']['texts']
tokenizer = Tokenizer(num_words=1000)  # à¸à¸³à¸«à¸™à¸”à¹ƒà¸«à¹‰à¹€à¸¥à¸·à¸­à¸à¸„à¸³à¸—à¸µà¹ˆà¹ƒà¸Šà¹‰à¸šà¹ˆà¸­à¸¢à¸ªà¸¸à¸” 1000 à¸„à¸³
tokenizer.fit_on_texts(texts_train)  # à¸ªà¸£à¹‰à¸²à¸‡à¸„à¸¥à¸±à¸‡à¸„à¸³à¸ˆà¸²à¸à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡

# à¹à¸›à¸¥à¸‡à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¹€à¸›à¹‡à¸™à¸¥à¸³à¸”à¸±à¸šà¸•à¸±à¸§à¹€à¸¥à¸‚
sequences = tokenizer.texts_to_sequences(texts_train)

# à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸¥à¸³à¸”à¸±à¸šà¸•à¸±à¸§à¹€à¸¥à¸‚à¸—à¸µà¹ˆà¹„à¸”à¹‰
print("Sequences (text to number):", sequences)

# Padding à¸¥à¸³à¸”à¸±à¸šà¸‚à¸­à¸‡à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¹ƒà¸«à¹‰à¸¡à¸µà¸„à¸§à¸²à¸¡à¸¢à¸²à¸§à¹€à¸—à¹ˆà¸²à¸à¸±à¸™
max_sequence_length = max(len(seq) for seq in sequences)
texts_padded = pad_sequences(sequences, maxlen=max_sequence_length)

print("Texts padded after padding:", texts_padded)

# à¸”à¸¶à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸„à¸¥à¸±à¸‡à¸„à¸³à¸­à¸­à¸à¸¡à¸²à¸”à¸¹ (word_index)
word_index = tokenizer.word_index
print("Word Index:", word_index)
print('texts to vector', texts_train)
```

##model

```python
import tensorflow as tf
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
from keras.utils import to_categorical
import numpy as np

# à¸•à¸±à¸§à¸­à¸¢à¹ˆà¸²à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥
texts_train = dataset['train']['texts']  # à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¹ƒà¸™à¸à¸²à¸£à¸à¸¶à¸
labels_train = dataset['train']['category']  # à¸›à¹‰à¸²à¸¢à¸à¸³à¸à¸±à¸šà¹ƒà¸™à¸à¸²à¸£à¸à¸¶à¸
texts_val = dataset['validation']['texts']  # à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¹ƒà¸™à¸à¸²à¸£ validation
labels_val = dataset['validation']['category']  # à¸›à¹‰à¸²à¸¢à¸à¸³à¸à¸±à¸šà¹ƒà¸™à¸à¸²à¸£ validation

# à¸‚à¸±à¹‰à¸™à¸•à¸­à¸™à¸—à¸µà¹ˆ 1: Tokenization
tokenizer = Tokenizer(num_words=1000)  # à¸ˆà¸³à¸à¸±à¸”à¸ˆà¸³à¸™à¸§à¸™à¸„à¸³à¸—à¸µà¹ˆà¹à¸•à¸à¸­à¸­à¸à¸¡à¸²à¹„à¸”à¹‰à¸ªà¸¹à¸‡à¸ªà¸¸à¸” 1000 à¸„à¸³
tokenizer.fit_on_texts(texts_train)

# à¹à¸›à¸¥à¸‡à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¹€à¸›à¹‡à¸™à¸¥à¸³à¸”à¸±à¸šà¸‚à¸­à¸‡à¸•à¸±à¸§à¹€à¸¥à¸‚ (sequences)
texts_train_sequences = tokenizer.texts_to_sequences(texts_train)
texts_val_sequences = tokenizer.texts_to_sequences(texts_val)

# à¸‚à¸±à¹‰à¸™à¸•à¸­à¸™à¸—à¸µà¹ˆ 2: Padding
max_sequence_length = max(len(seq) for seq in texts_train_sequences)  # à¸«à¸²à¸„à¸§à¸²à¸¡à¸¢à¸²à¸§à¸‚à¸­à¸‡à¸¥à¸³à¸”à¸±à¸šà¸—à¸µà¹ˆà¸¢à¸²à¸§à¸—à¸µà¹ˆà¸ªà¸¸à¸”
texts_train_padded = pad_sequences(texts_train_sequences, maxlen=max_sequence_length)
texts_val_padded = pad_sequences(texts_val_sequences, maxlen=max_sequence_length)

# label to one-hot encoded
num_classes = len(set(labels_train))  # label unique
labels_train = to_categorical(labels_train, num_classes=num_classes)
labels_val = to_categorical(labels_val, num_classes=num_classes)

model = Sequential()
model.add(Embedding(input_dim=1000, output_dim=100, input_length=max_sequence_length))
model.add(LSTM(128))
model.add(Dense(num_classes, activation='softmax'))  #  softmax for multi-class classification

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(texts_train_padded, labels_train, epochs=10, batch_size=32, validation_data=(texts_val_padded, labels_val))
```
