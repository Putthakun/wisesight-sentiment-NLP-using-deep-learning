##Step4: Data cleaning

```python
import re
import pandas as pd

def clean_text(example):
    text = example['texts']  # สมมติว่าคอลัมน์ข้อความคือ 'texts'

    # ลบ URLs
    text = re.sub(r'http\S+', '', text)

    # ลบอักขระที่ไม่ใช่ภาษาไทยและช่องว่าง
    text = re.sub(r'[^ก-๙\s]', '', text)

    # ลบช่องว่างซ้ำกันและลบช่องว่างต้น/ท้าย
    text = re.sub(r'\s+', ' ', text).strip()

    # ส่งกลับค่าที่แก้ไข
    example['texts'] = text
    return example

# ใช้ map กับ dataset ของคุณเพื่อทำความสะอาดข้อความในคอลัมน์ 'texts'
dataset['train'] = dataset['train'].map(clean_text)

#remove emoji
def remove_emojis(example):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emoticons
                               u"\U0001F300-\U0001F5FF"  # Symbols & Pictographs
                               u"\U0001F680-\U0001F6FF"  # Transport & Map Symbols
                               u"\U0001F1E0-\U0001F1FF"  # Flags
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251"  # Enclosed characters
                               "]+", flags=re.UNICODE)
    # แทนที่อิโมจิด้วยค่าว่าง
    example['train'] = emoji_pattern.sub(r'', example['train'])
    return example

# ฟังก์ชันลบช่องว่างซ้ำและส่งคืนเป็น dict
def remove_duplicate_spaces(example):
    text = example['texts']
    text = re.sub(r'\s+', ' ', text)  # แทนที่ช่องว่างหลายตัวด้วย 1 ช่องว่าง
    return {'texts': text.strip()}  # ตัดช่องว่างที่อยู่ต้นและท้ายข้อความ แล้วส่งคืนเป็น dict

# ใช้ map เพื่อแปลงข้อมูล
dataset['train'] = dataset['train'].map(remove_duplicate_spaces)
```

##Tokenization

```python
!pip install pythainlp
```

```python
from pythainlp.tokenize import word_tokenize

def tokenize_text(example):
    text = example['texts']  # สมมติว่าคอลัมน์ข้อความคือ 'texts'
    tokens = word_tokenize(text, engine='newmm')  # ใช้ newmm tokenizer ของ pythainlp
    example['texts'] = ' '.join(tokens)  # รวมคำที่ตัดแล้วให้เป็นประโยคใหม่
    return example

dataset['train'] = dataset['train'].map(tokenize_text)
```

##Normalization

```python
# ตัวอย่างการลบวรรณยุกต์
from pythainlp.util import normalize

def normalize_text(example):
    text = example['texts']  # สมมติว่าคอลัมน์ข้อความคือ 'texts'
    normalized_text = normalize(text)  # Normalize ข้อความให้เป็นมาตรฐาน
    return {'texts': normalized_text}  # ส่งคืนเป็น dict

dataset['train'] = dataset['train'].map(normalize_text)
```

```python
print(dataset['train'][0])
```

##Text Vectorization

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# สร้าง tokenizer และแปลงข้อความเป็นลำดับของตัวเลข

texts_train = dataset['train']['texts']
tokenizer = Tokenizer(num_words=1000)  # กำหนดให้เลือกคำที่ใช้บ่อยสุด 1000 คำ
tokenizer.fit_on_texts(texts_train)  # สร้างคลังคำจากข้อมูลข้อความ

# แปลงข้อความเป็นลำดับตัวเลข
sequences = tokenizer.texts_to_sequences(texts_train)

# ตรวจสอบลำดับตัวเลขที่ได้
print("Sequences (text to number):", sequences)

# Padding ลำดับของข้อความให้มีความยาวเท่ากัน
max_sequence_length = max(len(seq) for seq in sequences)
texts_padded = pad_sequences(sequences, maxlen=max_sequence_length)

print("Texts padded after padding:", texts_padded)

# ดึงข้อมูลคลังคำออกมาดู (word_index)
word_index = tokenizer.word_index
print("Word Index:", word_index)
print('texts to vector', texts_train)
```

##model

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
from keras.utils import to_categorical
import numpy as np

# ตัวอย่างข้อมูล
texts_train = dataset['train']['texts']  # ข้อความในการฝึก
labels_train = dataset['train']['category']  # ป้ายกำกับในการฝึก
texts_val = dataset['validation']['texts']  # ข้อความในการ validation
labels_val = dataset['validation']['category']  # ป้ายกำกับในการ validation

# ขั้นตอนที่ 1: Tokenization
tokenizer = Tokenizer(num_words=1000)  # จำกัดจำนวนคำที่แตกออกมาได้สูงสุด 1000 คำ
tokenizer.fit_on_texts(texts_train)

# แปลงข้อความเป็นลำดับของตัวเลข (sequences)
texts_train_sequences = tokenizer.texts_to_sequences(texts_train)
texts_val_sequences = tokenizer.texts_to_sequences(texts_val)

# ขั้นตอนที่ 2: Padding
max_sequence_length = max(len(seq) for seq in texts_train_sequences)  # หาความยาวของลำดับที่ยาวที่สุด
texts_train_padded = pad_sequences(texts_train_sequences, maxlen=max_sequence_length)
texts_val_padded = pad_sequences(texts_val_sequences, maxlen=max_sequence_length)

# label to one-hot encoded
num_classes = len(set(labels_train))  # label unique
labels_train = to_categorical(labels_train, num_classes=num_classes)
labels_val = to_categorical(labels_val, num_classes=num_classes)

model = Sequential()
model.add(Embedding(input_dim=1000, output_dim=100, input_length=max_sequence_length))
model.add(LSTM(128))
model.add(Dense(num_classes, activation='softmax'))  #  softmax for multi-class classification

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(texts_train_padded, labels_train, epochs=10, batch_size=32, validation_data=(texts_val_padded, labels_val))
```
